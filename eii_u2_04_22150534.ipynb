```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt


X = df[['Latitude',	'Longitude',	'Inventory_Level',	'Temperature']]
Y = df["Waiting_Time"]

# Datos de entrenamiento y datos de prueba

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de correlacion: {r2: 0.4f}\n')
```

```python
b0, b1, b2, b3, b4 = modelo.params

Fun = lambda x1, x2, x3, x4: b0 + b1 * x1 + b2 * x2 + b3 * x3 + b4 * x4

#si se trata de un modelo de predicción de demanda, un modelo de tiempo de espera
Fun(432, 60, 2, 120000)
```

CONCLUSION.

---


El valor P (PR(>F)) representa la probabilidad de obtener los resultados observados bajo la suposición de que la hipótesis nula es verdadera, es decir, que la variable independiente no tiene ningún efecto sobre la variable dependiente.

Cuando el valor P es inferior a 0.05, se considera que hay un efecto significativo. Sin embargo, en este caso, los valores P de todas las variables independientes (latitud, longitud, nivel de inventario y temperatura) son considerablemente mayores a 0.05. Esto indica que ninguna de estas variables influye de manera significativa en el tiempo de espera.
Por lo tanto, el modelo no logra predecir el tiempo de espera con precisión utilizando estas variables.
